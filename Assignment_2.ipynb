### Task 12-a.  Make a heatmap to show correlation among all the columns of the goal keeper dataframe and answer the questions

````python
#write code here to find the correlation

# Convert 'id' column to numeric, handling errors by setting non-convertible values to NaN
goalkeepers['id'] = pd.to_numeric(goalkeepers['id'], errors='coerce')

# Drop rows with NaN values in 'id' column
goalkeepers = goalkeepers.dropna(subset=['id'])

# Select only numeric columns for correlation calculation
numeric_goalkeepers = goalkeepers.select_dtypes(include=['number'])

# Calculate the correlation matrix for numeric columns only
corr_matrix = numeric_goalkeepers.corr()
````

````python
#write code here for heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
````
